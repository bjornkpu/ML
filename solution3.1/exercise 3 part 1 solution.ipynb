To do any further processing, we need to deal with the categorical variables. The simplest way is to do a one-hot encoding, where we create a vector of booleans for each possible value of a variable.

https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html

In pandas, we use the get_dummies() method.

```python
import matplotlib.pyplot as plt
import numpy as np

# Convert categoricals to dummies
Xt = pd.get_dummies(X)

print(X.shape)
print(Xt.shape)

# Plot document-term matrix, blue indicates non-zero value
plt.spy(Xt, markersize=1)
fig = plt.gcf()
fig.set_size_inches(50,400)
plt.plot()
plt.show()
```
